```python
def prob(y):
    return (1/(1+(math.e)**(-y)))
```

```python
def calc_y(x1,x2,b0,b1,b2):
    return b0+(x1*b1)+(x2*b2)
```

```python
def calc_b(x1,x2,b0,b1,b2,y,pred,alpha):
    b0=b0+(alpha*(y-pred)*pred*(1-pred)*1)
    b1=b1+(alpha*(y-pred)*pred*(1-pred)*x1)
    b2=b2+(alpha*(y-pred)*pred*(1-pred)*x2)
    return b0,b1,b2
```

```python
def normalize(X):
    return X-X.mean()
```
